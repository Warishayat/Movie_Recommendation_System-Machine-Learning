```python
def convert(obj):
  genres_name = []
  for i in ast.literal_eval(obj):
    genres_name.append(i["name"])
  return genres_name
```

```python
df["genres"]=df["genres"].apply(convert)
```

```python
df["keywords"][0]
```

```python
df["keywords"] = df["keywords"].apply(convert)
```

```python
def convert_cast(obj):
  genres_name = []
  counter = 0
  for i in ast.literal_eval(obj): #to get top 3 names from
    if counter !=3:
      genres_name.append(i["name"])
      counter+=1
    else:
      break
  return genres_name



df["cast"] = df["cast"].apply(convert_cast)
```

```python
def find_director(obj):
  director_name = []
  for i in ast.literal_eval(obj):
    if i["job"] == "Director":
      director_name.append(i["name"])
      break

  return director_name

df["crew"] = df["crew"].apply(find_director)
```

```python
df["overview"][0]
```

```python
df["overview"]=df["overview"].apply(lambda x:x.split())
```

```python
df["cast"] = df["cast"].apply(lambda x:[i.replace(" ","") for i in x])
df["crew"] = df["crew"].apply(lambda x:[i.replace(" ","") for i in x])
df["genres"] = df["genres"].apply(lambda x:[i.replace(" ","") for i in x])
df["keywords"] = df["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
```

# preprocessing_tags

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)

data["tags"] = data["tags"].apply(stem)
```

```python

vectors=cv.fit_transform(data["tags"]).toarray() #convert into array becasuse it return us a sparse matrix
```

```python
cv.get_feature_names_out()
```

```python
mostcommon_words=pd.DataFrame(cv.get_feature_names_out())
mostcommon_words.sample(30)
```

# CosineSimilarity For Recommendation

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape #we get 4806,4806 because we are finding similiarity from one moviw to each mobie inthe data
```

```python
similarity[1]   #movie metrics daigonal will be alywas 1 because movies are most similar with itself
```

```python
def recommend(movie):
    if movie not in data["title"].values:
        print(f"Movie '{movie}' not found in the dataset.")
        return

    movie_index = data[data["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:7]

    for key,value in enumerate(movies_list):
        print(f"The Smilar Movie from {movie} is:",data.iloc[value[0]]["title"])
```

```python
recommend("")
```
